# Analyse Random Forest

```python
import numpy as np
import time

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

start = time.time()
cancer = load_breast_cancer()


X_train, X_test, y_train, y_test = train_test_split(
cancer.data, cancer.target, stratify=cancer.target, train_size=0.7, random_state=66)

no_of_trees = 100 

clf = RandomForestClassifier(n_estimators=no_of_trees,random_state=0)

clf.fit(X_train, y_train)

score_test =  clf.score(X_test, y_test)
score_train = clf.score(X_train,y_train)

print('Random Forest (%i trees) test set score is %.3f'
      %(no_of_trees,score_test))
print('Random Forest (%i trees) training set score is %.3f'
      %(no_of_trees,score_train))
print('time elapsed: %.3f [s]' % (time.time()-start))
```

## Plot feature importance

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
n_features = cancer.data.shape[1]
plt.barh(range(n_features), clf.feature_importances_,align='center')
plt.yticks(np.arange(n_features), cancer.feature_names)
plt.xlabel('attribute/feature importance')
plt.ylabel('attribute name')
plt.show()
```

# Analyse gradient boosted decision tree (GBDT)

```python
import numpy as np
import time

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier

start = time.time()
cancer = load_breast_cancer()


X_train, X_test, y_train, y_test = train_test_split(
cancer.data, cancer.target, random_state=0)

no_of_trees = 1000 
learning_rate = 0.01
tree_depth = 1

clf = GradientBoostingClassifier(n_estimators=no_of_trees,
                                 learning_rate=learning_rate,
                                 max_depth=tree_depth,
                                 random_state=0)

clf.fit(X_train, y_train)

score_test =  clf.score(X_test, y_test)
score_train = clf.score(X_train,y_train)

print('GBDT (%i trees) test set score is %.3f'
      %(no_of_trees,score_test))
print('GBDT (%i trees) training set score is %.3f'
      %(no_of_trees,score_train))
print('time elapsed: %.3f [s]' % (time.time()-start))
```

## Plot feature importance

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
n_features = cancer.data.shape[1]
plt.barh(range(n_features), clf.feature_importances_,align='center')
plt.yticks(np.arange(n_features), cancer.feature_names)
plt.xlabel('attribute/feature importance')
plt.ylabel('attribute name')
plt.show()
```
